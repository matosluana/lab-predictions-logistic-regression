### Importing and querying the complete data

```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn import neighbors
```

```python
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
```

```python
query = '''SELECT r.rental_id, r.rental_date, p.amount, f.title, f.rental_duration, f.rental_rate, f.length, f.rating, 
f.special_features, c.name AS 'category'
FROM sakila.rental r
RIGHT JOIN sakila.inventory i USING (inventory_id)
LEFT JOIN sakila.film f USING (film_id)
LEFT JOIN sakila.payment p USING (rental_id)
LEFT JOIN sakila.film_category fc USING (film_id)
LEFT JOIN sakila.category c USING (category_id)
LEFT JOIN sakila.language l USING (language_id)'''

data = pd.read_sql_query(query, engine)
display(data.head())
data.shape
```

### Checking whether the movies have been rented in May 2005 or not

```python
def check_rental(text):
        if '2005-05' in str(text):
            return 'rented'
        else:
            return 'not rented'

        
data["rental_status"] = data['rental_date'].apply(lambda x: check_rental(x))
data
```

### Data cleaning

```python
# Now that we have the rental status we do not need rental id and rental date anymore

data = data.drop(columns=['rental_id', 'rental_date'], axis=1)
data
```

```python
data.isnull().sum()
#I will replace those with zero
```

```python
data = data.fillna(0)
data.isnull().sum()
```

```python
# If the same movie has multiple copies but only one has been rented, the model will be negatively influenced.
# I will drop all movies copies which have not been rented, if there is a copy of the same title which was rented.

data = data.sort_values('rental_status', ascending=False)
data
```

```python
data = data.drop_duplicates(subset='title', keep="first")
data
```

```python
data['rental_status'].value_counts()
```

```python
# I have to split the special features

#Get all unique values of special features

special_features = data['special_features'].unique() #all possible values for features
sf = [] #list will be filled with separated features

for text in special_features:
    features = re.findall("[^,]+",text) # get all individual features separated by comma
    for feature in features:   #adding those to the list of features
        if feature in sf:
            pass
        else:
            sf.append(feature)
display(sf)      

#adding the features to the table (already with "one hot encoding")
for feature in sf:
    data[feature]=data['special_features'].apply(lambda x: 1 if feature in x else 0)


data = data.drop(columns=['special_features'], axis=1)
data
```

### X-y split, train test split, scaling and encoding

```python
X = data.drop(columns=['rental_status'], axis=1)
y = data ['rental_status']

display(X)
display(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
# Scaling the numerical train data

X_train_num = X_train.select_dtypes(include = np.number)
transformer = MinMaxScaler().fit(X_train_num) # need to keep transformer
X_train_normalized = transformer.transform(X_train_num)
X_train_norm = pd.DataFrame(X_train_normalized)
X_train_norm.columns = X_train_num.columns
X_train_norm.head()
```

```python
#checking correlations

corr_matrix=data.corr(method='pearson')  # default
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
#Checking the distributions

sns.displot(X_train_norm['amount'])


sns.displot(X_train_norm['rental_duration'])


sns.displot(X_train_norm['rental_rate'])


sns.displot(X_train_norm['length'])
```

```python
# Encoding the train categoricals

X_train_categorical = X_train.select_dtypes(include = object)
X_train_categorical = X_train_categorical.drop(columns=['title'])
X_train_cat = pd.get_dummies(X_train_categorical, 
                             columns=['rating','category'],
                             drop_first=True)
X_train_cat.head()
```

```python
X_train_transformed = np.concatenate([X_train_norm, X_train_cat], axis=1)
```

```python
# Scaling the numerical test data

X_test_num = X_test.select_dtypes(include = np.number)
# we use the transformer that was trained on the training data
X_test_normalized = transformer.transform(X_test_num)
X_test_norm = pd.DataFrame(X_test_normalized)
```

```python
# Encoding the categorical test data

X_test_categorical = X_test.select_dtypes(include = object)
X_test_categorical = X_test_categorical.drop(columns=['title'])
X_test_cat = pd.get_dummies(X_test_categorical, 
                            columns=['rating','category'],
                            drop_first=True)
# verify that dummies columns are in the same order and that the same column was dropped
display(list(zip(list(X_train_cat.columns),list(X_test_cat.columns))))
# not needed if you treat each dataframe with one_hot_encoder and save the encode (and the column names)

X_test_cat.head()
```

```python
X_test_transformed = np.concatenate([X_test_norm, X_test_cat], axis=1)
```

### Building and testing the model

##### Using logistic regression

```python
classification = LogisticRegression(random_state=0, solver='saga',
                  multi_class='multinomial').fit(X_train_transformed, y_train)
```

```python
predictions = classification.predict(X_test_transformed)
classification.score(X_test_transformed, y_test)
```

#### Using KNN

```python
clf = neighbors.KNeighborsClassifier(n_neighbors=5, weights='uniform')
clf.fit(X_train_transformed, y_train)
predictions_clf = clf.predict(X_test_transformed)
clf.score(X_test_transformed, y_test)
```

#### Comparing predictions from both models

```python
print(y_test.value_counts())
```

```python
pd.Series(predictions_clf).value_counts() #KNN
```

```python
confusion_matrix(y_test, predictions)
```
